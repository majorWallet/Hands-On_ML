## 1. From biological neurons to artificial neurons

### 1-1. biological neuron

**Neurons** consist of a cell body and complex components containing a **nucleus**. There is a branch-shaped projection called **dendrite** and a very long projection called **axon**. **Axons** can be several to 40,000 times longer than cell bodies. The ends of the **axon** are divided into **telodendria**, which are microscopic structures called synaptic terminals and are connected to **dendrites** or cell bodies of other neurons. Biological neurons create short electrical impulses called action potentials or simply signals. This signal travels along the **axon**, causing the synapse to generate a chemical signal called **neurotransmitter**. **Neurons** generate their own signals when they receive a sufficient amount of **neurotransmitter** for a few thousand of a second.

### 1-2. logical operation using neuron

The **artificial neuron** model has one or more binary inputs and one binary output. The **artificial neuron** simply exports the output when the input is activated by a certain number.


![img](artificial_neuron.png)


- What this network does
    - The first network on the left is an identity function. When neuron A is activated, neuron C is also activated. But when neuron A is turned off, neuron C is turned off.
    - The second network performs logical multiplication operations. Neuron C is activated only when both neurons A and B are activated.
    - A third network performs a logical sum operation. When either neuron A or B is activated, neuron C is activated.
    - In the fourth network, neuron C is activated when neuron A is activated and neuron B is deactivated.

### 1-3. perceptron

- A perceptionron is called a TLU (threshold logic unit) or a LTU (linear threshold unit).
- The input and output are any number other than binary, and each input connection is related to a weight.
- The TLU calculates the weight sum of the inputs(z = w_1x_1 + w_2x_2 + ... + w_nx_n = w^Tx) and applies a step function to the calculated sum to output the result.
    - step function: h_w(x) = step(z)


![img](perceptron.png)


- The most widely used step function in perceptron is the Heaviside step function.
- Sometimes they use the sign function instead.

- A perceptron consists of a single layer of TLU
- When all neurons in a layer are connected to all neurons in the previous layer, it is called a fully connected layer or a dense layer.
